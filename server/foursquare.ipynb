```python
import foursquare
import requests
from pprint import pprint
import time
from fsKeys import clientid, secret
import pymongo


# Mongo connection information
conn = 'mongodb://localhost:27017'
mongo_client = pymongo.MongoClient(conn)
db = mongo_client.foursquare

# Starting point of the FourSquare query for coffee shops
fs_client = foursquare.Foursquare(client_id=clientid, client_secret=secret)
zipcode_search=[91932, 92014, 92025, 92037, 92071, 92101, 92102, 92103, 92104, 92105, 92106, 92107, 92108, 92109, 92110, 92111, 92113, 92114, 92115, 92116, 92117, 92119, 92120, 92121, 92123, 92124, 92126, 92127, 92128, 92130, 92131, 92139, 92154, 92173]
```

```python
for zc in zipcode_search:
    response = fs_client.venues.search(params={'query': 'coffee', 'near': zc, 'limit': 50})

    venues=[]
    for x in response["venues"]:
        if("address" in x["location"]):
            venue={}
            venue["city"]=""
            venue["state"]=""
            venue["zip"]=""
            venue["createdDate"]=""
            venue["fsid"] = x["id"]
            venue["name"] = x["name"]
            venue["lat"] = x["location"]["lat"]
            venue["lng"] = x["location"]["lng"]
            venue["address"] = x["location"]["address"]
            if("city" in x["location"]):
                venue["city"] = x["location"]["city"]
            if("state" in x["location"]):
                venue["state"] = x["location"]["state"]
            if("zip" in x["location"]):
                venue["zip"] = x["location"]["postalCode"]
            venues.append(venue)
    coffee = db.coffeeshops
    for x in venues:
        if(coffee.find_one({"fsid":x["fsid"]})):
            print("found "+str(x["fsid"])+" in the coffeeshops table")
        else:
            print("did not find "+str(x["fsid"])+" in the coffeeshops table")
            db.coffeeshops.insert_one(
                {
                    "fsid":x["fsid"],
                    "name":x["name"],
                    "address":x["address"],
                    "city":x["city"],
                    "state":x["state"],
                    "zip":x["zip"],
                    "lat":x["lat"],
                    "lng":x["lng"],
                    "createdDate":x["createdDate"]
                }
            )
            print("added "+str(x["fsid"])+" to the coffeeshops table")
```

```python
coffeeShops = db.coffeeshops.find()
for shop in coffeeShops:
    if(shop["createdDate"] == ""):
        try:
            details = fs_client.venues(shop["fsid"])
            createdDate = time.strftime('%m/%d/%Y',  time.gmtime(details["venue"]["createdAt"]))
            db.coffeeshops.update_one(
                {"fsid":shop["fsid"]},
                {
                "$set":{
                    "createdDate":createdDate
                }
                }
            )
            print("record updated successfully")
        except Exception as e:
            print(e)
```

```python
# coffeeShops = db.coffeeshops.find({"createdDate"},{"name":1, "address":0, "city":0, "state":0, "zip":0, "lat":1, "lng":1, "createdDate":1})
import json
import datetime

coffeeShops = db.coffeeshops.find()
output={"type":"FeatureCollection"}
features = []
count=0

for shop in coffeeShops:
    feature={}
    properties={}
    geometry={}
    
    if(shop["createdDate"] != ""):
        olddt=datetime.datetime.strptime(shop["createdDate"], '%m/%d/%Y')
        
        feature["type"]="Feature"
        
        properties["time"] = olddt.strftime('%Y-%m-%d') 
        feature["properties"]=properties
        
        geometry["type"]="Point"
        geometry["coordinates"]=[shop["lng"],shop["lat"],1]
        feature["geometry"]=geometry
        
        features.append(feature)
# print(features)
output["features"]=features
print(json.dumps(output))
```
